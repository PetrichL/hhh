Сделаем категории **"Вид ДТП"** и **"Улица"** раноговыми 

```python
df_train=df_train.merge(pd.DataFrame(df_train.groupby(['id'])['id'].count()),left_on='id', right_index=True,suffixes=('', '_count'))
df_test=df_test.merge(pd.DataFrame(df_test.groupby(['id'])['id'].count()),left_on='id', right_index=True,suffixes=('', '_count'))

df_test
```

```python
#статистика по гексагонам h3 на тестовой выборке
h3_3=pd.DataFrame(df_train.groupby(['h3-3'])['id'].count())
h3_4=pd.DataFrame(df_train.groupby(['h3-4'])['id'].count())
h3_5=pd.DataFrame(df_train.groupby(['h3-5'])['id'].count())
h3_6=pd.DataFrame(df_train.groupby(['h3-6'])['id'].count())
h3_7=pd.DataFrame(df_train.groupby(['h3-7'])['id'].count())
h3_8=pd.DataFrame(df_train.groupby(['h3-8'])['id'].count())
h3_8
```

```python

df_train=df_train.merge(h3_3,left_on='h3-3', right_index=True,suffixes=('', '_h3_3'))
df_train=df_train.merge(h3_4,left_on='h3-4', right_index=True,suffixes=('', '_h3_4'))
df_train=df_train.merge(h3_5,left_on='h3-5', right_index=True,suffixes=('', '_h3_5'))
df_train=df_train.merge(h3_6,left_on='h3-6', right_index=True,suffixes=('', '_h3_6'))
df_train=df_train.merge(h3_7,left_on='h3-7', right_index=True,suffixes=('', '_h3_7'))
df_train=df_train.merge(h3_8,left_on='h3-8', right_index=True,suffixes=('', '_h3_8'))

```

```python
df_test=df_test.merge(h3_3,how="left",left_on='h3-3', right_index=True,suffixes=('', '_h3_3')).fillna(0)
df_test=df_test.merge(h3_4,how="left",left_on='h3-4', right_index=True,suffixes=('', '_h3_4')).fillna(0)
df_test=df_test.merge(h3_5,how="left",left_on='h3-5', right_index=True,suffixes=('', '_h3_5')).fillna(0)
df_test=df_test.merge(h3_6,how="left",left_on='h3-6', right_index=True,suffixes=('', '_h3_6')).fillna(0)
df_test=df_test.merge(h3_7,how="left",left_on='h3-7', right_index=True,suffixes=('', '_h3_7')).fillna(0)
df_test=df_test.merge(h3_8,how="left",left_on='h3-8', right_index=True,suffixes=('', '_h3_8')).fillna(0)
```

```python
df_test[['Место','Улица','Дом','Дорога']]=df_test[['Место','Улица','Дом','Дорога']].fillna('')
df_train[['Место','Улица','Дом','Дорога']]=df_train[['Место','Улица','Дом','Дорога']].fillna('')

df_test=df_test.fillna(-1)
df_train=df_train.fillna(-1)
```

Пострим на распределение данных

```python
X = df_train.drop(["Дата", "Время","Место","Дорога", "Дом", "id", "Погибло",	"Погибло детей",	"Ранено",	"Ранено детей"], axis = 1)
y = df_train[["Погибло",	"Погибло детей",	"Ранено",	"Ранено детей"]]

df_test=df_test.drop(["Дата", "Время","Место","Дорога", "Дом"], axis = 1)
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
y_train
```

## Обучение модели

```python
from catboost import CatBoostRegressor
from catboost import CatBoostClassifier
```

```python
cat_features=['Улица', 'Вид ДТП','h3-3', 'h3-4','h3-5', 'h3-6', 'h3-7', 'h3-8']
```

```python
y_train['Погибло'][y_train['Погибло']>0]=1
y_train['Погибло детей'][y_train['Погибло детей']>0]=1
y_train['Ранено'][y_train['Ранено']>0]=1
y_train['Ранено детей'][y_train['Ранено детей']>0]=1

y_test['Погибло'][y_test['Погибло']>0]=1
y_test['Погибло детей'][y_test['Погибло детей']>0]=1
y_test['Ранено'][y_test['Ранено']>0]=1
y_test['Ранено детей'][y_test['Ранено детей']>0]=1
```

```python
model1 = CatBoostClassifier(iterations=1000,
                          depth=12)
#,eval_metric='BalancedAccuracy'
model1.fit(X_train, y_train['Погибло'],
          cat_features=cat_features,use_best_model=True,early_stopping_rounds=100,
          eval_set=(X_test, y_test['Погибло']),
         plot=True,verbose=False)
```

```python
model2 = CatBoostClassifier(iterations=1000,
                          depth=12)
#,eval_metric='BalancedAccuracy'
model2.fit(X_train, y_train['Погибло детей'],
          cat_features=cat_features,use_best_model=True,early_stopping_rounds=100,
          eval_set=(X_test, y_test['Погибло детей']),
         plot=True,verbose=False)
```

```python
model3 = CatBoostClassifier(iterations=1000,
                          depth=12)
#,eval_metric='BalancedAccuracy'
model3.fit(X_train, y_train['Ранено'],
          cat_features=cat_features,use_best_model=True,early_stopping_rounds=100,
          eval_set=(X_test, y_test['Ранено']),
         plot=True,verbose=False)
```

```python
model4 = CatBoostClassifier(iterations=1000,
                          depth=12)
#,eval_metric='BalancedAccuracy'
model4.fit(X_train, y_train['Ранено детей'],
          cat_features=cat_features,use_best_model=True,early_stopping_rounds=100,
          eval_set=(X_test, y_test['Ранено детей']),
         plot=True,verbose=False)
```
